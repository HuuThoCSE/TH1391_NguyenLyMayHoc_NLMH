**import thư viện**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import keras
from keras.models import Sequential
from keras.layers import Dense
```

**Load dữ liệu**

```python
dataset = pd.read_csv('/content/drive/MyDrive/NLMH_Thuchanh/Week12_Thuchanh6/Churn_Modelling.csv')
dataset.head()
```

```python
X = dataset.iloc[:, 3:13].values
y = dataset.iloc[:, 13].values
```

**Chuẩn hoá dữ liệu**

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X_1 = LabelEncoder()
X[:, 1] = labelencoder_X_1.fit_transform(X[:, 1])
labelencoder_X_2 = LabelEncoder()
X[:, 2] = labelencoder_X_2.fit_transform(X[:, 2])
onehotencoder = OneHotEncoder() #nếu báo lỗi thì đổi thành OneHotEncoder()
X = onehotencoder.fit_transform(X).toarray()
X = X[:, 1:]
```

**Tách dữ liệu làm train và test**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

**Chuẩn hoá đặc trưng**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

**Xây dựng ANN và huấn luyện**

```python
# classifier = Sequential()
# classifier.add(Dense(output_dim = 6, init = 'uniform', activation = 'relu', input_dim = 11))
# classifier.add(Dense(output_dim = 6, init = 'uniform', activation = 'relu'))
# classifier.add(Dense(output_dim = 1, init = 'uniform', activation = 'sigmoid'))
# # Compiling the ANN
# classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
# # Fitting the ANN to the Training set
# classifier.fit(X_train, y_train, batch_size = 10, nb_epoch = 100)
```

```python
classifier = Sequential()
classifier.add(Dense(6, kernel_initializer = 'uniform', activation = 'relu', input_dim=16934))
classifier.add(Dense(6, kernel_initializer = 'uniform', activation = 'relu'))
classifier.add(Dense(1, kernel_initializer = 'uniform', activation = 'sigmoid'))
# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
# Fitting the ANN to the Training set
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

**Hiển thị độ chính xác mô hình**

```python
loss, accuracy = classifier.evaluate(X_train, y_train, verbose=0)
print('Accuracy: %.2f' % (accuracy*100))
```

**Kiểm thử mô hình**

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
for i in range(5):
 print('%s => %d (expected %d)' % (X_test[i].tolist(), y_pred[i],y_test[i]))
```

**Hiển thị ma trận nhầm lẫn**

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
f, ax = plt.subplots(figsize=(8,5))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt=".0f", ax=ax)
plt.xlabel("y_pred")
plt.ylabel("y_true")
plt.show()
```
