```python
import pandas as pd
import numpy as np
abalone = pd.read_csv("/content/drive/MyDrive/TH1391_NLMH/Week10_Thuchanh4/abalone.data")
```

**Kiểm tra dữ liệu**

```python
abalone.head()
```

**Xác định name columns**

```python
abalone.columns = ["Sex","Length","Diameter","Height","Whole weight","Shucked weight","Viscera weight","Shell weight","Rings",]
```

**Xóa dữ liệu dư thừa**

Mục tiêu của bài tập hiện tại là sử dụng các phép đo vật lý để dự đoán tuổi của bào ngư. Vì giới tính không phải là thước đo thuần túy về thể chất nên xóa nó khỏi tập dữ liệu.

```python
abalone = abalone.drop("Sex", axis=1)
```

**Tìm K Nearest Neighbors**

```python
X = abalone.drop("Rings", axis=1)
X = X.values
y = abalone["Rings"]
y = y.values
```

**Khởi tạo 1 point mới**

```python
new_data_point = np.array([0.569552,0.446407,0.154437,1.016849,0.439051,0.222526,0.291208,])
```

**Sử dụng hàm linalg.norm() để tính khoảng cách giữa hai điểm dựa trên Euclid**

```python
distances = np.linalg.norm(X - new_data_point, axis=1)
```

**Chọn K = 3 và sử dụng argsort() để sắp xếp mảng từ thấp đến cao**

```python
k = 3
nearest_neighbor_ids = distances.argsort()[:k]
nearest_neighbor_ids
```

**Xác định vị trí của các khoảng cách vừa tìm được**

```python
nearest_neighbor_rings = y[nearest_neighbor_ids]
nearest_neighbor_rings
```
