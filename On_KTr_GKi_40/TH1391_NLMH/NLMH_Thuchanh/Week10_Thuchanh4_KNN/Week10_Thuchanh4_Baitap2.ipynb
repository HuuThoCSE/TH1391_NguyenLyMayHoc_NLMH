```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.impute import SimpleImputer

# Tải dữ liệu
penguins = pd.read_csv('/content/drive/MyDrive/TH1391_NLMH/Week10_Thuchanh4/penguins.csv')

# Drop rows where target is NaN
penguins = penguins.dropna(subset=['species'])  # replace 'species' with your actual target column

# Tiền xử lý dữ liệu
features = penguins.select_dtypes(include=[np.number])
target = penguins['species']  # replace 'species' with your actual target column

# Fill NaN values with mean of the respective column
imputer = SimpleImputer(strategy='mean')
features = imputer.fit_transform(features)

# Convert target to numerical
le = LabelEncoder()
target = le.fit_transform(target)

# Tách dữ liệu
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Khởi tạo và huấn luyện mô hình
knn = KNeighborsClassifier(n_neighbors=5, p=2)
knn.fit(X_train, y_train)

# Dự đoán và đánh giá
y_pred = knn.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
