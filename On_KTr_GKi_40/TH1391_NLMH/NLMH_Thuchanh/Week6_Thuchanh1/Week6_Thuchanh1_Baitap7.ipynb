```python
import pandas as pd
```

```python
df = pd.DataFrame(
    [
        [180000, 110, 18.9, 1400],
        [360000, 905, 23.4, 1800],
        [230000, 230, 14.0, 1300],
        [60000, 450, 13.5, 1500]
    ],
    columns=["Feature 1", "Feature 2", "Feature 3", "Feature 4"]
)
df
```

**7.1:** Sử dụng chuẩn hóa min - max (Sử dụng hàm và hàm tự dựng)

```python
# Ham
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result
```

```python
normalized_df = normalize(df)
print(normalized_df)
```

```python
# Cong thuc
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(df)
print(normalized_data)
```

```python
normalized_df = pd.DataFrame(normalized_data, columns=df.columns)
print(normalized_df)
```

**7.2:** Sử dụng chuẩn hóa z-score (Sử dụng hàm và hàm tự dựng)

```python
### Tu dung
def z_score_normalize_df(df, columns):
    normalized_df = df.copy()
    for col in columns:
        mean = df[col].mean()
        std_dev = df[col].std()
        normalized_df[col] = (df[col] - mean) / std_dev
    return normalized_df
```

```python
# Chọn các cột muốn chuẩn hóa
columns_to_normalize = ["Feature 1", "Feature 2", "Feature 3", "Feature 4"]
```

```python
# Chuẩn hóa dữ liệu
normalized_df = z_score_normalize_df(df, columns_to_normalize)
print("Kết quả tự xây dựng hàm:")
print(normalized_df)
```

```python
### Ham
import scipy.stats as stats
# Chuẩn hóa dữ liệu bằng z-score
zscore_df = df.apply(stats.zscore)
print(zscore_df)
```
