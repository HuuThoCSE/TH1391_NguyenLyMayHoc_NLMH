**Bài tập 4:** Xây dựng hàm chuẩn hóa Z-score dựa theo công thức với dữ liệu df. Đối chiếu với kết quả khi sử dụng hàm.


```python
import pandas as pd
```

```python
df = pd.DataFrame.from_dict({
    'Name': ['Nik', 'Kate', 'Joe', 'Mitch', 'Alana'],
    'Age': [32, 30, 67, 34, 20],
    'Income': [80000, 90000, 45000, 23000, 12000],
    'Education' : [5, 7, 3, 4, 4]
})
print(df.head())
```

```python
def z_score_normalize_df(df, columns):
    normalized_df = df.copy()
    for col in columns:
        mean = df[col].mean()
        std_dev = df[col].std()
        normalized_df[col] = (df[col] - mean) / std_dev
    return normalized_df
```

```python
# Chọn các cột muốn chuẩn hóa
columns_to_normalize = ['Income'] # columns_to_normalize = ['Age', 'Income', 'Education']

# Chuẩn hóa dữ liệu
normalized_df = z_score_normalize_df(df, columns_to_normalize)
print("Kết quả tự xây dựng hàm:")
print(normalized_df)
```

```python
from scipy.stats import zscore
```

```python
z_scores = zscore(df)
```

```python
z_scores_df = pd.DataFrame(z_scores, columns=df.columns)
print(z_scores_df)
```

**7.3:** Sử dụng chuẩn hóa thang thập phân

```python
import numpy as np
```

```python
def decimal_scaling(column):
    max_abs_value = np.max(np.abs(column))
    k = np.ceil(np.log10(max_abs_value))
    return column / (10 ** k)
```
