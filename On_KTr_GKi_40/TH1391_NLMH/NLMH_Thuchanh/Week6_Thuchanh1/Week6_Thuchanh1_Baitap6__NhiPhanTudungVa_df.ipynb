```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/TH1391_NLMH/Week6_Thuchanh1/california_housing_train.csv")
df
```

```python
# Chuyá»ƒn Ä‘á»•i DataFrame thÃ nh máº£ng numpy
data_array = df.values

# TÃ¬m giÃ¡ trá»‹ tuyá»‡t Ä‘á»‘i lá»›n nháº¥t
max_abs_value = np.max(np.abs(data_array))

# TÃ­nh sá»‘ chá»¯ sá»‘ cá»§a giÃ¡ trá»‹ tuyá»‡t Ä‘á»‘i lá»›n nháº¥t
k = np.ceil(np.log10(max_abs_value))

# Chuáº©n hÃ³a dá»¯ liá»‡u vá»›i thang tháº­p phÃ¢n
decimal_scaled_data = data_array / (10 ** k)

# Chuyá»ƒn Ä‘á»•i máº£ng numpy Ä‘Ã£ chuáº©n hÃ³a thÃ nh DataFrame
decimal_scaled_df = pd.DataFrame(decimal_scaled_data, columns=df.columns)

# Hiá»ƒn thá»‹ DataFrame Ä‘Ã£ chuáº©n hÃ³a
print(decimal_scaled_df)
```
