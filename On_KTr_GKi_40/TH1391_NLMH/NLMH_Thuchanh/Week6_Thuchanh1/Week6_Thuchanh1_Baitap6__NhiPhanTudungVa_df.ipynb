```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/TH1391_NLMH/Week6_Thuchanh1/california_housing_train.csv")
df
```

```python
# Chuyển đổi DataFrame thành mảng numpy
data_array = df.values

# Tìm giá trị tuyệt đối lớn nhất
max_abs_value = np.max(np.abs(data_array))

# Tính số chữ số của giá trị tuyệt đối lớn nhất
k = np.ceil(np.log10(max_abs_value))

# Chuẩn hóa dữ liệu với thang thập phân
decimal_scaled_data = data_array / (10 ** k)

# Chuyển đổi mảng numpy đã chuẩn hóa thành DataFrame
decimal_scaled_df = pd.DataFrame(decimal_scaled_data, columns=df.columns)

# Hiển thị DataFrame đã chuẩn hóa
print(decimal_scaled_df)
```
