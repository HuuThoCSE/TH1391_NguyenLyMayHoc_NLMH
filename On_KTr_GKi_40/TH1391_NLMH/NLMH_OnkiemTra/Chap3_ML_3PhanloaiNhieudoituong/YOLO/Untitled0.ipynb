```python
import os
import xml.etree.ElementTree as ET

# Function to convert XML to YOLO format
def convert_xml2yolo(file_path, classes):
    with open(file_path, 'r') as in_file:
        txt_outfile = file_path.replace('.xml', '.txt')
        tree = ET.parse(in_file)
        root = tree.getroot()
        out_file = open(txt_outfile, 'w')
        for obj in root.iter('object'):
            cls = obj.find('name').text
            if cls not in classes:
                continue
            cls_id = classes.index(cls)
            xmlbox = obj.find('bndbox')
            b = (int(xmlbox.find('xmin').text), int(xmlbox.find('ymin').text), int(xmlbox.find('xmax').text), int(xmlbox.find('ymax').text))
            out_file.write(str(cls_id) + " " + " ".join([str(a) for a in b]) + '\n')
        out_file.close()

# Define your classes
classes = ['orange']

# Convert all XML files in a specific directory
for file in os.listdir('/content/orange'):  # Change this path to your XML files directory
    if file.endswith('.xml'):
        convert_xml2yolo(os.path.join('/content/orange', file), classes)  # Change this path to your XML files directory

# After converting XML to YOLO format, you can proceed with the training of YOLO model as in the provided guide.
```
