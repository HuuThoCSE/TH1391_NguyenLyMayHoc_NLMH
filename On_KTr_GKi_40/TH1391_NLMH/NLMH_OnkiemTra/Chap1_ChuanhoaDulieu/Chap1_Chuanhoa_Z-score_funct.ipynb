# **Chuẩn hóa Z-score với dataframe**

```python
# Loading a Sample Pandas Dataframe
import pandas as pd
from scipy import stats
import numpy as np

df = pd.DataFrame.from_dict({
    'Name': ['Nik', 'Kate', 'Joe', 'Mitch', 'Alana'],
    'Age': [32, 30, 67, 34, 20],
    'Income': [80000, 90000, 45000, 23000, 12000],
    'Education' : [5, 7, 3, 4, 4]
})

print(df.head())
```

**Chuẩn hóa Z-score**

```python
# Áp dụng một cột cụ thể
df['Income zscore'] = stats.zscore(df['Income'])
print(df.head())
```

**Chuyển đổi các cột trong khung dữ liệu của mình thành Z-score sử dụng hàm apply()trong Pandas**

```python
df = df.select_dtypes(include='number').apply(stats.zscore)
print(df.head())
```

```python
# Áp dụng toàn bộ cột
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[numeric_cols] = df[numeric_cols].apply(stats.zscore)
```
