```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/NLMH_OnkiemTra/chap1_california_housing_train.csv")
df
```

```python
import numpy as np

def decimal_scaling(column):
    max_abs_value = np.max(np.abs(column))
    k = np.ceil(np.log10(max_abs_value))
    return column / (10 ** k)

# Tìm các cột dữ liệu số
numerical_columns = df.select_dtypes(include=np.number).columns

# Chuẩn hóa các cột dữ liệu số với thang thập phân
df[numerical_columns] = df[numerical_columns].apply(decimal_scaling)
```
