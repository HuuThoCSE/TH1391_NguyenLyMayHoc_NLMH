```python
# Loading a Sample Pandas Dataframe
import pandas as pd

df = pd.DataFrame.from_dict({
    'Name': ['Nik', 'Kate', 'Joe', 'Mitch', 'Alana'],
    'Age': [32, 30, 67, 34, 20],
    'Income': [80000, 90000, 45000, 23000, 12000],
    'Education' : [5, 7, 3, 4, 4]
})

print(df.head())
```

```python
# Chuẩn hóa một cột
def zscore_col(df, column_name):
    mean = df[column_name].mean()
    std_dev = df[column_name].std(ddof=0) # sử dụng độ lệch chuẩn dân số
    df[column_name] = (df[column_name] - mean) / std_dev
    return df

def zscore(df):
    for column in df.select_dtypes(include=['float64', 'int64']).columns:
        df = zscore_col(df, column)
    return df
```

```python
# Chọn các cột muốn chuẩn hóa
col = ['Income'] # columns_to_normalize = ['Age', 'Income', 'Education']

# Chuẩn hóa dữ liệu
df_zscore_col = zscore_col(df, col)
print("Kết quả tự xây dựng hàm:")
print(df_zscore_col)
```

```python
# Chuẩn hóa dữ liệu
df_zscore = zscore(df)
print("Kết quả tự xây dựng hàm:")
print(df_zscore)
```
