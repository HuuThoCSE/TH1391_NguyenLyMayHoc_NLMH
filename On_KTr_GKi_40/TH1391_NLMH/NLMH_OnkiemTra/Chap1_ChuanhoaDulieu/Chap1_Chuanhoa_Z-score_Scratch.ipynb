import thư viện

```python
import numpy as np
import scipy.stats as stats
```

Khởi tạo mảng giá trị

```python
data = np.array([6, 7, 7, 12, 13, 13, 15, 16, 19, 22])
```

Chuẩn hóa Z-score - Kết quả:

[-1.394, -1.195, -1.195, -0.199, 0, 0, 0.398, 0.598, 1.195, 1.793]

Giá trị đầu tiên của “6” trong mảng là 1,394 độ lệch chuẩn dưới giá trị trung bình.

Giá trị thứ năm của “13” trong mảng là 0 độ lệch chuẩn so với giá trị trung bình, tức là nó bằng giá trị trung bình.

Giá trị cuối cùng của “22” trong mảng là 1,793 độ lệch chuẩn trên giá trị trung bình.

```python
zscores = stats.zscore(data)

# Lấy hai chữ số sau dấu phẩy cho mỗi giá trị z-score - NEW
rounded_zscores = np.round(zscores, 3)

print(rounded_zscores)
```
