# **Sử dụng chuẩn hóa min - max (Sử dụng hàm và hàm tự dựng)**

```python
from sklearn.preprocessing import MinMaxScaler

# Chuẩn hóa dữ liệu
scaler = MinMaxScaler()
scaler_data = scaler.fit_transform(df)

scaler_data_df = pd.DataFrame(scaler_data, columns=df.columns)
scaler_data_df
```

```python
def scaler_normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result

scaler_math_data_df = scaler_normalize(df)
print(scaler_math_data_df)
```

# **Sử dụng chuẩn hóa z-score (Sử dụng hàm và hàm tự dựng)**

```python
from scipy.stats import zscore

# Áp dụng toàn bộ cột
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
zscore_data_df = df[numeric_cols].apply(zscore)

zscore_data_df.head()
```

```python
# Chuẩn hóa toàn bộ cột
def zscore_math(df):
    for column_name in df.select_dtypes(include=['float64', 'int64']).columns:
        mean = df[column_name].mean()
        std_dev = df[column_name].std(ddof=0) # sử dụng độ lệch chuẩn dân số
        df[column_name] = (df[column_name] - mean) / std_dev
    return df

zscore_math_df = zscore_math(df)
print("Kết quả tự xây dựng hàm:")
zscore_math_df
```

```python
import numpy as np

def decimal_scaling(column):
    max_abs_value = np.max(np.abs(column))
    k = np.ceil(np.log10(max_abs_value))
    return column / (10 ** k)

# Tìm các cột dữ liệu số
numerical_columns = df.select_dtypes(include=np.number).columns

# Chuẩn hóa các cột dữ liệu số với thang thập phân
decimal_math_df = df[numerical_columns].apply(decimal_scaling)

print("Dữ liệu sau khi chuẩn hóa với thang thập phân:")
decimal_math_df.head()
```
