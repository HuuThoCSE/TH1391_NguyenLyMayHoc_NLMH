```python
# Bài tập 1: Xây dựng hàm chuẩn hóa MinMax theo công thức và thực hiện chuẩn hóa cho mảng dữ liệu data. Đối chiếu với kết quả khi sử dụng hàm MinMaxScaler
```

```python
import pandas as pd
```

```python
data = {'weight':[300, 250, 800],
        'price':[3, 2, 5]}
df = pd.DataFrame(data)
print(df)
```

```python
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result

normalized_df = normalize(df)
print(normalized_df)
```
